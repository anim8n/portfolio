```python
import os
from PIL import Image

# 获取当前路径下的所有子文件夹
def get_immediate_subdirectories(directory):
    return [os.path.join(directory, d) for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]

# 遍历每个子文件夹，查找封面图片并压缩清晰度，另存为 _compress 版本
def compress_covers_in_subdirs(directory):
    subdirs = get_immediate_subdirectories(directory)
    for subdir in subdirs:
        for root, dirs, files in os.walk(subdir):
            for file in files:
                if file in ["封面.png", "封面.jpg"]:
                    file_path = os.path.join(root, file)
                    name, ext = os.path.splitext(file)
                    new_file_name = f"{name}_compress{ext}"
                    new_file_path = os.path.join(root, new_file_name)
                    try:
                        with Image.open(file_path) as img:
                            # 如果是 JPEG，使用 quality 参数压缩清晰度
                            if file.lower().endswith((".jpg", ".jpeg")):
                                img.save(new_file_path, quality=30, optimize=True)
                            # 如果是 PNG，可以降低颜色深度或使用 optimize
                            elif file.lower().endswith(".png"):
                                img = img.convert("P", palette=Image.ADAPTIVE, colors=64)  # 降低颜色到32色
                                img.save(new_file_path, optimize=True)
                            print(f"Compressed and saved: {new_file_path}")
                    except Exception as e:
                        print(f"Error processing {file_path}: {e}")

# 调用函数，只处理当前路径下的子文件夹
compress_covers_in_subdirs(".")
```
